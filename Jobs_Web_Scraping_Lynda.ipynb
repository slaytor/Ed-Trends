```python
from robobrowser import RoboBrowser
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from lxml import html

import datetime
today = datetime.date.today().strftime('%Y-%m-%d')
```

# Lynda

```python
base = 'https://www.lynda.com'

page = requests.get('https://www.lynda.com/Home/Authors.aspx')
soup = BeautifulSoup((page).content, 'html.parser')

authors = [x.find('a').get('href') for x in list(soup.findAll('div', {'class': 'details-row'}))]
#authors = authors[:2]

frames = []

for author in authors:
    print(author)
    page = requests.get(base + author)
    soup = BeautifulSoup((page).content, 'html.parser')
    courses = [x for x in list(soup.findAll('div', {'class': "card card-list-style search-result course"}))]
    
    author_dict = {'title': [],
               'url': [],
               'author': [],
               'duration': [],
               'level': [],
               'released': [],
               'views': [],
               'subjects': [],
               'description': [],
               'skills': []}

    for course in courses:
        title = course.find('h3').text
        url = course.find('a', {'class': 'ga ga-multiple'}).get('href')
    
        if course.find('cite'):
            author = course.find('cite').text
        else:
            author = 'None'
        
        if course.find('span', {'class': 'meta-duration'}):
            duration = course.find('span', {'class': 'meta-duration'}).text
        else:
            duration = 'None'
    
        if course.find('span', {'class': 'meta-level'}):
            level = course.find('span', {'class': 'meta-level'}).text
        else:
            level = 'None'
    
        if course.find('span', {'class': 'release-date'}):
            released = course.find('span', {'class': 'release-date'}).text
        else:
            released = 'None'
        
        if course.find('span', {'class': 'meta-views'}):
            views = course.find('span', {'class': 'meta-views'}).text
        else:
            views = 'None'
                
    
        author_dict['title'].append(title)
        author_dict['url'].append(url)
        author_dict['author'].append(author)
        author_dict['duration'].append(duration)
        author_dict['level'].append(level)
        author_dict['released'].append(released)
        author_dict['views'].append(views)
        
        
        page = requests.get(url)
        soup = BeautifulSoup((page).content, 'html.parser')
        
        if soup.find('div', {'typeof': "BreadcrumbList"}):
            subjects = [x.find('span').text for x in soup.findAll('div', {'typeof': "BreadcrumbList"})]
        else:
            subjects = 'None'
        
        if soup.find('div', {'itemprop': "description"}):
            description = soup.find('div', {'itemprop': "description"}).text
        else:
            description = 'None'
            
        if soup.findAll('em'):
            skills = [x.text for x in soup.findAll('em')]
        else:
            skills = 'None'
    
        author_dict['subjects'].append(subjects)
        author_dict['description'].append(description)
        author_dict['skills'].append(skills)
    
    author_df = pd.DataFrame(author_dict)
    frames.append(author_df)
    
lynda_df = pd.concat(frames)
lynda_df.to_csv('data/jobs/lynda/lynda_' +today+ '.csv', index=False)
```

```python
df = pd.concat(frames)
```
